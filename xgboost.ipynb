# 数据预处理，划分训练集和测试集

```python
#提取特征矩阵和标签
x=train_df.iloc[:,1:-1]
y=train_df.iloc[:,-1]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

#切分训练集和测试集,并且进行随机数打乱
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,test_size=0.2,random_state=12)
```

# 训练xgboost模型

```python
seed=512
```

```python
        xgb_params = {
            'booster': 'gbtree',
            'objective': 'binary:logistic',
            'n_estimators': 200,
            'max_depth': 8,
            'lambda': 10,
            'subsample': 0.7,
            'colsample_bytree': 0.8,
            'colsample_bylevel': 0.7,
            'eta': 0.1,
            'tree_method': 'hist',
            'seed': seed,
            'nthread': 16
        }

```

```python
# 您可以从X_train和y_train中划分一部分作为验证集
Xval = Xtrain.sample(frac=0.2, random_state=12)  # 10%的X_train作为验证集
Yval = Ytrain.loc[Xval.index]  # 对应的y_train的子集
```

```python
from xgboost import XGBClassifier
import xgboost as xgb
# 训练XGBoost模型
xgb_model = xgb.XGBClassifier(**xgb_params, eval_metric='auc',early_stopping_rounds=20)
xgb_model.fit(Xtrain, Ytrain,eval_set=[(Xtrain,Ytrain),(Xval, Yval)],verbose=True)

```

# 计算auc数值

```python
# 预测测试集
Ypred = xgb_model.predict_proba(Xtest)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
# 评估模型性能
test_auc = roc_auc_score(Ytest, Ypred)
print(f"Test AUC: {test_auc}")
```

# 绘制出ROC曲线

```python
# 绘制测试集的ROC曲线
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
fpr, tpr, thresholds = roc_curve(Ytest, Ypred)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % test_auc)
plt.plot([0, 1], [0, 1], 'k--')  # 随机猜测的ROC曲线
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Test Set')
plt.legend(loc="lower right")
plt.show()
```
